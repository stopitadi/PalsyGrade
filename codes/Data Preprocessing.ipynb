```python
from PIL import Image
import os

def resize_images(input_folder, output_folder, target_size=(100, 100)):
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    # List all image files in the input folder
    image_files = [f for f in os.listdir(input_folder) if f.endswith('.jpg') or f.endswith('.jpeg') or f.endswith('.png')]
    
    # Process each image and save the resized output
    for image_file in image_files:
        # Generate input and output paths
        input_path = os.path.join(input_folder, image_file)
        output_path = os.path.join(output_folder, image_file)
        
        # Open and resize the image
        try:
            with Image.open(input_path) as img:
                resized_img = img.resize(target_size, Image.ANTIALIAS)
                resized_img.save(output_path)
                print(f"Resized {image_file} and saved to {output_path}")
        except Exception as e:
            print(f"Error processing {image_file}: {e}")

# Example usage:
input_folder = r"C:\Users\hp\Desktop\detection of the bels palsy\mediapipe\AffectedCr_non"  # Replace with your input folder path
output_folder = r"C:\Users\hp\Desktop\detection of the bels palsy\mediapipe\AffectedCr_non"  # Replace with your output folder path

# Resize images in the input folder and save to output folder
resize_images(input_folder, output_folder)
```

# MODEL

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten
```

```python
model = Sequential([
    Conv2D(32, (3,3), activation = 'relu', input_shape = (100, 100, 3)), 
    MaxPooling2D((2,2)), 

    Conv2D(32, (3,3), activation='relu'),
    MaxPooling2D((2,2)), 

    Flatten(), 
    Dense(64, activation = 'relu'), 
    Dense(1, activation= 'sigmoid')
    
])
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
model.fit(x_train, y_train, epochs = 5, batch_size = 64 )
```

```python
model.evaluate(x_test, y_test)
```

# data augmentation 

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam

# Example CNN model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(100, 100, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(32, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Compile the model
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

# Data augmentation using ImageDataGenerator
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    vertical_flip=False
)
```

```python
# Fit the model with data augmentation
history = model.fit(datagen.flow(x_train, y_train, batch_size=64),
                    ),  # Convert to integer
                    epochs=5)
```

```python
model.evaluate(x_test, y_test)
```
